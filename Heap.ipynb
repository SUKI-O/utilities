## Python 3 Implimentation of Heap DS.


```python
# Heap is represented in an indexed Array
class HEAP:
    def __init__(self, arr):
        self.type = None
        self.maxsize = 1000
        self.size = len(arr)
        self.Heap = [0]*(self.maxsize+1)
        self.Heap[1:len(arr)+1] = arr
        self.root = 1
    
    def parent(self, i):
        if i == self.root:
            return None
        return i//2
    
    def left(self, i):
        if 2*i <= self.size:
            return 2*i
        else:
            return None
    
    def right(self, i):
        if 2*i+1 <= self.size:
            return 2*i+1
        else:
            return None
        
    def swap(self, i, j):
        self.Heap[i], self.Heap[j] = self.Heap[j], self.Heap[i]
        
    def max_Heapify(self, i):
        left = self.left(i)
        right = self.right(i)

        if left != None and self.Heap[left] > self.Heap[i]:
            largest = left
        else:
            largest = i
        if right != None and self.Heap[right] > self.Heap[largest]:
            largest = right
        if largest != i:
            self.swap(i, largest)
            self.max_Heapify(largest)

    def build_max_Heap(self):
        self.type = "max"
        for i in range(self.size//2, 0, -1):
            self.max_Heapify(i)

    def min_Heapify(self, i):
        left = self.left(i)
        right = self.right(i)

        if left != None and self.Heap[left] < self.Heap[i]:
            smallest = left
        else:
            smallest = i
        if right != None and self.Heap[right] < self.Heap[smallest]:
            smallest = right
        if smallest != i:
            self.swap(i, smallest)
            self.min_Heapify(smallest)
    
    def build_min_Heap(self):
        self.type = "min"
        for i in range(self.size//2, 0, -1):
            self.min_Heapify(i)
    
    def print_heap(self):
        print([i for i in self.Heap[self.root:self.size+1]])
    
    def extract_root(self):
        if self.size <1:
            print('heap underflow')
            return
        root = self.Heap[self.root]
        self.size = self.size - 1
        self.Heap[self.root]= self.Heap[self.size+1]
        if self.type == "max":
            self.max_Heapify(1)
        if self.type == "min":
            self.min_Heapify(1)
        return root
        
    def increase_key(self, i, key):
        if self.type == "max":
            self.Heap[i] = key
            while i >1 and self.Heap[self.parent(i)] < self.Heap[i]:
                    self.swap(i, self.parent(i))
                    i = self.parent(i)
                    
        if self.type == "min":
            self.Heap[i] = key
            while i >1 and self.Heap[self.parent(i)] > self.Heap[i]:
                    self.swap(i, self.parent(i))
                    i = self.parent(i)

    def insert(self, key):
        import math
        self.size +=1
        if self.type == "max":
            self.Heap[self.size+1]== - (math.inf)
            self.increase_key(self.size, key)

        if self.type == "min":
            self.Heap[self.size+1]== math.inf
            self.increase_key(self.size, key)

        
```

```python
arr = [10, 8, 2, 4, 6, 3, 8, 20, 28, 1, 5]
heap = HEAP(arr)
heap.build_min_Heap()
heap.print_heap()
heap.insert(-1)
heap.print_heap()
heap.insert(22)
heap.print_heap()
heap.insert(50)
heap.print_heap()
for i in range(heap.size):
    print(heap.extract_root(), end=' ')
```
